```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
from csv import writer
import numpy as np
import pandas as pd
import json
import matplotlib
from matplotlib import ticker as ticker
import matplotlib.pyplot as plt
import h5py
```

# Visualizes the cache trace for target and non-target (For Targeted Deanonymization Data)

```python
#Change the filename here (ends in .json)

filename ="flipStress_p12_t0.5s.json"
data = json.load(open(filename))
trainX = np.array(data['X'],dtype=np.float32)
trainY=np.array(data['y'])
print(trainX.shape)
print(trainY.shape)

fig = plt.figure(figsize=(15, 5))

# Prints nth target data and nth non-target datapoint (0-99 are target, 100-199 are non target)
plt.plot(trainX[0], label='Target',  color='blue', linewidth=1.5, alpha=0.5)
plt.plot(trainX[100], label='Non-Target',  color='orange', linewidth=1.5, alpha=0.5)
plt.legend()

# Set X-axis ticks and labels to represent time in seconds
num_ticks = 5000 // 500  # Calculate the number of ticks
plt.xticks(np.arange(0, 5500, 500), labels=np.arange(0, num_ticks + 1))


plt.xlabel("Time (s)")
plt.ylabel("Buffer access time")
plt.grid()

plt.savefig("output_plot.pdf", format="pdf", dpi=300)
plt.show()
```

# Visualize cache trace for a website (website fingerprining data)

```python
#Enter filename here (ends in .csv)
filename="chase1729729198453.csv"

# Load the CSV file
WF = pd.read_csv(filename, header=None)
WF = WF.to_numpy()

# Calculate time in seconds (5ms per unit, so divide by 1000 to get seconds)
time = [i * 5 / 1000 for i in range(WF.shape[1])]

# Plot the data
fig = plt.figure(figsize=(15, 5))
plt.plot(time, WF[0], label=filename)


plt.xlabel('Time (s)')
plt.ylabel('Buffer Access Tim e')
plt.legend()
plt.grid()
plt.show()
```
